**Let's read from the csv file** `grades.csv`

````python


file = open('sampleData/grades.csv', 'r')

lines = [] #The list we'll fill with each line from the file
for line in file:
    lines.append(line)
    #Uncomment this to visually verify the lines being imported
    #print(line)
file.close()
````

Let's create summaries of these values and put them in a new file

````python
#We'll separate each line into its values and put those in a list
data_table = []
for line in lines:
    stripped_line = line.strip() #Removes \n from the end of the line
    list_of_values = stripped_line.split(",")
    #Let's see what's in the list, just to check:
    print(list_of_values)
    data_table.append(list_of_values)
````

````python
#Define an average function
def average(lst):
    sum = 0
    length = len(lst)
    for val in lst:
        sum += int(val)
    return sum / length
````

````python

output_file_name = "averages.txt"

out_file = open(output_file_name, "w") #We need to write to this file
````

````python



for row in data_table:
    output_text = row[0] #the student id
    output_text += ": "
    output_text += str(average(row[1:]))
    output_text += "\n" # Don't forget to add the newline!
    print("Writing text to file:", output_text)
    out_file.write(output_text)
    
````

````python
out_file.close() #Very important!
````
