````python
#lists in lists:

chessboard = [["r","n","b","q","k","b","n","r",],
              ["p","p","p","p","p","p","p","p",],
              [" "," "," "," "," "," "," "," ",],
              [" "," "," "," "," "," "," "," ",],
              [" "," "," "," "," "," "," "," ",],
              [" "," "," "," "," "," "," "," ",],
              ["P","P","P","P","P","P","P","P",],
              ["R","N","B","Q","K","B","N","R",],]


def print_chessboard(board):
    for row_num in range(len(board)):
        print("DEBUG: Current Row:", board[row_num])
        for col_num in range(len(board[row_num])):
            if board[row_num][col_num] != " ":
                print("Piece:", board[row_num][col_num], "at position", str(row_num + 1) + "," + str(col_num+1))

print_chessboard(chessboard)
````
