**scopingDemo.py**

````python
import math

'''
var_x = 10

def do_something_global():
    print("Var x value:", var_x)
    
def do_something_local():
    var_x = 20
    print("Var x local value:", var_x)
    
def really_bad_idea():
    global var_x
    var_x = 30
    print("Global var_x modified locally:", var_x)

do_something_global()
    
do_something_local()

print("Var x before bad idea:", var_x)
really_bad_idea()
print("Var x after bad idea:", var_x)
'''


def func():
    var_x = 100
    print("var_x = ", var_x)
    # print(locals()) #Prints the local variables in this function




def square(x):
    '''
    squares a number and returns the result
    '''
    
    result = x * x
    return result

def calculate_hypotenuse(side_a, side_b):
    """
    given two sides of a right triangle, calculates the
    length of the longest side (the hypotenuse)
    parameters:
      side_a: One side
      side_b: the other side
    returns:
        The hypotenuse of the triangle
    """
    
    side_c_squared = square(side_a) + square(side_b)
    hypotenuse = math.sqrt(side_c_squared)
    return hypotenuse
    

def main():
    var_x = 89

    print("var_x = ", var_x)
    func()
    print("var_x = ", var_x)


some_global_var = "Hi everybody!"

if __name__ == "__main__":
    main()
else:
    print("Loaded as module named:", __name__)
````
