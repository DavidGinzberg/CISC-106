**list_exercise_answers.py**

````python
# Exercise:
import cisc106

# 1) Write a function that returns the second-last element
# of a list passed to it
def second_last(some_list):
    # Need to use the subscript notation (some_list[index])
    #x = len(some_list)
    #return some_list[x - 2]
    return some_list[-2] #another way -- these are both valid options

#tests for second_last:
cisc106.assertEqual(second_last([1, 2, "red", "blue"]), "red")
cisc106.assertEqual(second_last([1, 2, ["red", "0xFF0000"], "blue"]), ["red", "0xFF0000"])

# 2) Write a function that checks whether a list is empty
# and return True if it is, and False if it isn't
def is_empty(some_list):
    if type(some_list) != type([]):
        return False
    return len(some_list) == 0
    
    
cisc106.assertEqual(is_empty([]), True)
cisc106.assertEqual(is_empty(["text"]), False)
cisc106.assertEqual(is_empty(42), False)
    

````
