**vacation_demo.py**

````python
#start_day = 0   length_vacation = 10    end_day = 3
#start_day = 3   length_vacation = 10    end_day = 6
#start_day = 6   length_vacation = 23    end_day = 1

import cisc106


#function header here
def vacation_return_date(start_day, length_vacation):
    '''Calculates return day of week based on start day and length of
    vacation
    Arguments:
        start_day - The day of the week (0-6) you left for vacation
        length_vacation - How many nights you're gone
    returns:
        end_day - day of the week (0-6) when you return from vacation
    '''
    # Formula: (start + length) % 7 -> end
    return (start_day + length_vacation) % 7 #on planets where there are 7 days in a week

def main():
    # 1. ask the user for the starting day -- As an *int*
    starting_day = int(input("What day are you leaving for vacation?"))

    # 2. ask the user for the number of days on vacation -- As an *int*
    days_gone = int(input("How long are you gone?"))

    # 3. use formula to compute the ending day
    end_date = vacation_return_date(starting_day, days_gone)

    # 4. display the ending day to the screen
    print("You are returning on day number:", end_date)


def run_tests():
    # Formula: (start + length) % 7 -> end

    # Simplest case:
    #start_day = 0   length_vacation = 1
    # end_day = (start_day plus length_vacation) -> 0+1 -> 1
    cisc106.assertEqual(vacation_return_date(0, 1), 1)

    #start_day = 6   length_vacation = 3
    # end_day = (start_day plus length_vacation) -> 6+3 -> 9 (should be: 2)
    # (better formula): end_day = (start + length) mod 7 (% sign)
    cisc106.assertEqual(vacation_return_date(6, 3), 2)


    #start_day = 0   length_vacation = 10    end_day = 3
    cisc106.assertEqual(vacation_return_date(0, 10), 3)
    #start_day = 3   length_vacation = 10    end_day = 6
    cisc106.assertEqual(vacation_return_date(3, 10), 6)
    #start_day = 6   length_vacation = 23    end_day = 1
    cisc106.assertEqual(vacation_return_date(6, 23), 1)


if __name__ == "__main__":
    main()
````

````python
# While loops --
#    iteration: lets you run a block of code any number of times automatically
#

# Structure of a while loop:
# The `while` keyword
# followed by a condition
# Followed by a block of code

#Example:

num = 1
# Stops before the value 10 is printed
while num < 10:
    print("Current number:", num)
    num = num + 1

#This won't run because the condition is already false
while num > 1000:
    print("We should never see this print because num is always < 1000")
    
countdown = 10
while countdown >= 0:
    #print(f"T-minus {countdown} seconds") #is the same as:
    print("T-minus", countdown, "seconds") # Which is the same as:
    #print("T-minus " + str(countdown) + " seconds")
    countdown = countdown - 1
    

# Exercise:
'''
Write a guessing game: Ask a player for a number, and tell them if they guessed
a random number (1-10) correctly

V1: Start with guessing the number once

V2: keep guessing until the player gets it right
'''
#To generate a random number use random.randrange(start, end)
#examples:
import random
print(random.randrange(4, 17))
````
