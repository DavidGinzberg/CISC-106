````python
'''Functions for working with lists, including answers to exercises
'''
import cisc106 #for testing functions

def contains_empty_list(some_list):
    '''
    returns true if the given list contains an empty list
    '''
    
    return [] in some_list

def is_empty_list(some_list):
    return type(some_list) == type([]) and len(some_list) == 0


# Exercise:
# Create a function that takes a list and some value as two arguments
# And returns true if that list does NOT contain the value stored in the
# second argument
def does_not_contain(some_list, missing_element):
    return missing_element not in some_list
        
    
cisc106.assertEqual(does_not_contain([], "text"), True)
cisc106.assertEqual(does_not_contain([5], 5), False)
cisc106.assertEqual(does_not_contain([42], [42]), True)
cisc106.assertEqual(does_not_contain([1, 2, [42], 100], [42]), False)


'''
End of class question:
values = ['one', 'two’, 3, 'four’, -5]

Which expression evaluates to True?

values[4] > values[2]
‘two’ == values[2]
[‘four’] not in values
len(values) == 4
'''
````
