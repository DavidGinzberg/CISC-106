**midtermdemo.py:**

````python
"""
Ask the user to input a midterm 1 grade and a
midterm 2 grade, then print the higher of the two
"""

# Number of inputs: 2
# What do we do with those inputs: Assign them to variables
# Output: Which one is bigger (print that one)

midterm_grade_1 = input("Midterm Grade 1: ")
midterm_grade_2 = input("Midterm Grade 2: ")

midterm_grade_1 = float(midterm_grade_1)
midterm_grade_2 = float(midterm_grade_2)

print("Your highest grade was: ", max(midterm_grade_1, midterm_grade_2))
````

**parkingspotdemo.py:**

````python
last_name = "Montoya"
spot_number = 45

def parking_pass(spot, name):
    if type(name) == type("") and type(spot) == type(25):
        return f"Parking pass for {name}. Parking is reserved in spot number {spot}"
    else:
        return "Sorry, one of your variables was the wrong type"
````
