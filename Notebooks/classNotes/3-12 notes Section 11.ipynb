**vacation_function.py**

````python
'''
This is an implementation of the following problem:
data - type:    start_day :  int
            length_vacation :  int
            end_day :   int
test cases:      start_day = 0   length_vacation = 10    end_day = 3
              start_day = 3   length_vacation = 10    end_day = 6
              start_day = 6   length_vacation = 23    end_day = 1
'''

import cisc106

# Steps to solve:
def main():
    # 1. ask the user for the starting day -- As an int
    start_day = int(input("What day are you leaving for vacation? (0-6)"))
    # 2. ask the user for the number of days on vacation -- as an int
    days_gone = int(input("How many days will you be gone?"))

    # 3. use formula to compute the ending day (use our function)
    return_day = vacation_end_day(start_day, days_gone)

    # 4. display the ending day to the screen
    print("You will be back on day: ", return_day)

    # Extra credit: write a function that converts the return day number
    # to a day of the week (eg: 1 -> "Monday")

def vacation_end_day(start_day, length_vacation):
    #pass #lets us write a function that does nothing but can be called in tests
    #return start_day + length_vacation # Well that kind of worked... try this:
    return (start_day + length_vacation) % 7 

#These are our tests

#we'll call our function vacation_end_day(...)
#start_day = 2   length_vacation = 3    end_day = 5
cisc106.assertEqual(vacation_end_day(2, 3), 5)
#start_day = 0   length_vacation = 10    end_day = 3
cisc106.assertEqual(vacation_end_day(0, 10), 3)
#start_day = 3   length_vacation = 10    end_day = 6
cisc106.assertEqual(vacation_end_day(3, 10), 6)
#start_day = 6   length_vacation = 23    end_day = 1
cisc106.assertEqual(vacation_end_day(6, 23), 1)

# Remember how to create a main function:
if __name__ == "__main__":
    main()
````
