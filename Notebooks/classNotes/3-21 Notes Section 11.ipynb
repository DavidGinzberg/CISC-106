**sample_list_answers.py**

````python
import cisc106
# function that changes the elements in a list
# This is called a "side-effect"
def decrement_first(some_list):
    '''
    decrement_first reduces the value of the first element
    of a list by 3
    arguments:
        some_list -- the list to be changed
    return: None
    '''
    some_list[0] = some_list[0] - 3
    
    
# In-class exercise:
# Create a function that takes a list as an argument
# And replaces its first element with "hello"
# and its last element with "goodbye"
# and returns the result

# Create two versions -- one that changes the original list, one that
# creates a copy and leaves the original alone
def change_first_and_last(a_list):
    a_list[0] = "hello"
    a_list[-1] = "goodbye"
    return a_list

def change_first_and_last_no_side_effect(a_list):
    # Use a slice to make a copy:
    copy_list = a_list[:] #no start and no end means it takes the whole list
    copy_list[0] = "hello"
    copy_list[-1] = "goodbye"
    return copy_list #return the new list that has the changed values

## in class:
# create a function that takes two lists as arguments
# and compares those lists
# Return "Aliases" if they are aliases of the same list
# Return "Different" if they are different"
# Return "Same Contents" if they are two separate lists, that are not aliases
def compare_lists(list_one, list_two):
    ''' Compare two lists
        Return:
            "Aliases" if they are aliases of the same list
            "Different" if they are different"
            "Same Contents" if they are two separate lists, that are not aliases
    '''
    if list_one is list_two: #this means they're aliases
        return "Aliases"
    if list_one == list_two: #same contents, but not aliases
        return "Same Contents"
    else:
        return "Different"
    
test_list = [1,2,3]
test_alias = test_list
test_copy = test_list[:]
cisc106.assertEqual(compare_lists(test_list, test_alias), "Aliases")
cisc106.assertEqual(compare_lists(test_list, test_copy), "Same Contents")
cisc106.assertEqual(compare_lists(test_list, [5,6,7]), "Different")
# How do we test these?
#remember to import cisc106 (at the top of this file)
#Test for change_first_and_last
test_list = ["Apple", "Orange", "Microsoft", "Verizon"]
expected_result = ["hello", "Orange", "Microsoft", "goodbye"]
cisc106.assertEqual(change_first_and_last(test_list), expected_result)
cisc106.assertEqual(test_list[0], "hello")
cisc106.assertEqual(test_list[-1], "goodbye")
                    
#test for change_first_and_last_no_side_effect
test_list = [1,2,3,4] #reset test list
test_list_start = test_list[0] #the original value in index 0
test_list_end = test_list[-1] # the original value at the end of the list
expected_result = ["hello", 2, 3, "goodbye"]
cisc106.assertEqual(change_first_and_last_no_side_effect(test_list), expected_result)
cisc106.assertEqual(test_list[0],test_list_start) #test that the list didn't change
cisc106.assertEqual(test_list[-1], test_list_end)



# We could improve this file by putting all the tests in a `main` function
````
