**functions_calling_answers.py**

````python
import cisc106 #this should be at the top of the file

def func_a(some_list):
    #this is an example of slicing
    print("ignoring start of list:", some_list[:-1])
    return some_list[-1]

def func_b(a, b, c):
    list_of_nums = [a, b, c]
    print("list of nums:", list_of_nums)
    return func_a(list_of_nums)

def increment_last_element(a_list):
    a_list[-1] = a_list[-1] + 1
    #no return, we're calling this for it's side effect of changing the list


# in class exercise:
# Create a function that takes a list, and one string, and returns a new
# list with the first and last elements changed to that string
# NOTE: The original list should remain the same
# example: if we have my_list = [1,2,3] and we call
# new_list = replace_first_and_last(my_list, "hello")
# new_list should be ["hello", 2, "hello"]
def replace_first_and_last(a_list, replacement):
    copy_list = a_list[:] #not an alias
    copy_list[0] = replacement
    copy_list[-1] = replacement
    return copy_list

def replace_first_and_last_with_alias(a_list, replacement):
    copy_list = a_list #This is an alias because we just use `=` without a slice
    copy_list[0] = replacement
    copy_list[-1] = replacement
    return copy_list

# in class exercise:
# Create function that takes two lists, and returns a string saying:
# "The same list" if they are aliases
# "Different" if they contain different values
# "Equivalent copies" if they contain the same values but are not aliases
def compare_lists(list_one, list_two):
    if list_one is list_two:
        return "The same list"
    if list_one == list_two:
        return "Equivalent copies"
    else:
        return "Different"
        


# Tests for replace_first_and_last
test_list = [1,2,3]
cisc106.assertEqual(replace_first_and_last(test_list, "hello"), ["hello", 2, "hello"])
cisc106.assertEqual(test_list[0], 1) #should still be the same
cisc106.assertEqual(test_list[-1], 3) #should still be the same

#Tests for compare_lists
test_list_alias = test_list
test_list_clone = test_list[:]
test_list_2 = [3,2,1]
cisc106.assertEqual(compare_lists(test_list, test_list_alias), "The same list")
cisc106.assertEqual(compare_lists(test_list, test_list_clone), "Equivalent copies")
cisc106.assertEqual(compare_lists(test_list, test_list_2), "Different")


print(func_b(5,7,9)) #

````
