**midterm_review1.py**

````python
def mystery(x):
    if x < 5: # Block 1
        return  x  -  4 
    elif x == 5: #Block 2
        return x
    else: #Block 3
        return x + 4

def main():
    samp  =  36  //  7
    #print(mystery(samp))
    result = mystery(samp)
    print("Result was:", result)

main()
````
