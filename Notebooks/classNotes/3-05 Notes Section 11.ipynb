**variable_scope_demo.py**

````python
door_height = 84

print("Height of the doorway:", door_height, "inches")

def printDoorStats():
    '''Prints stats about a door'''
    
    print("The height of the door (defined globally):", door_height, "inches")
    door_width = 48
    print("The width of the door (defined locally):", door_width, "inches")
    return

some_variable = 10

def funky_function():
    global some_variable
    some_variable = 25
    print("Local&global value of some_variable in funky_function:", some_variable)
    
print("Variable before calling function:", some_variable)
funky_function()
print("Variable after calling function:", some_variable)
    
def print_local_shadow_var():
    some_variable = 42
    print("The answer is:", some_variable)
    return

print("Variable before shadowing:", some_variable)
print_local_shadow_var()
print("Variable after shadowing:", some_variable)

def the_right_way_to_use_global_variables(some_argument):
    """Prints a variable that was passed as an argument"""
    print("Before changing:", some_argument)
    some_argument = 15
    print("After changing:", some_argument)
    return

print("Variable before passing to function:", some_variable)
the_right_way_to_use_global_variables(some_variable)
print("Variable after passing to function:", some_variable)

````

**geometry.py**

````python
import math
import cisc106

def square(x):
    '''returns the square of a number'''
    
    result = x ** 2
    return result

def find_hypotenuse(base, height):
    '''uses pythagoras' theorem (a^2 + b^2 = c^2) to calculate
    the hypotenuse of a right triangle
    arguments:
        base -- base of triangle (side a)
        height -- height of triangle (side b)
    returns:
        hypotenuse (side_c)
    '''
    # The right way, for the notes
    #c_squared = square(base) + square(height)
    #hypotenuse = math.sqrt(c_squared)
    #return hypotenuse
    c_squared = square(base) + square(height)
    #hypotenuse = math.sqrt(c_squared)
    return c_squared
    
def main():
    """ The testing function for geometry.py """
    cisc106.assertEqual(find_hypotenuse(3,4), 5.0)
    cisc106.assertEqual(find_hypotenuse(15,24), 28.3019)
    cisc106.assertEqual(square(2), 4)
    cisc106.assertEqual(square(10), 100)
    return


if __name__ == "__main__":
    main()
````
