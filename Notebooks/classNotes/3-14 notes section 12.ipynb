**flow_of_execution2.py**

````python
def mystery(x):
    if x > 5 : #True
        print("a") #This runs (and prints 'a')
    elif x == 6 : #skipped -- This can never run
        print("b") #skipped
    if x == 6: #True
        print("c") # This runs (and prints 'c')
    elif x > 3: #Skipped because the `if` ran
        print("d") #skipped
    else: #won't run
        print("e") #skipped
    return "f" #Runs, returns 'f', and ends the function

def main():
    b = 6
    print(mystery(b))

main()

````

**flow_of_execution2-5.py**

````python
def mystery(x):
    if x > 5 :
        return "a" #if this runs, function stops and we only get 'a'
    elif x == 6 :
        print("b")
    if x == 6:
        print("c")
    elif x > 3:
        print("d")
    else:
        print("e")
    return "f"

def main():
    b = 6
    print(mystery(b))
    c = 5
    #d = mystery(c)
    print(f"Return value of calling mystery({c}): {mystery(c)}")

main()
````
