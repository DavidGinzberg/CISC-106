**math_functions.py**

````python
import cisc106

def square(factor):
    """
    Square function multiplies a number by itself
    arguments:
        factor (float,int,clowns)
    returns:
        factor ^ 2 (float)
    """
    result = factor * factor
    # result = factor ** 2
    return result


def computeRectangleArea(width, height):
    '''
    Computes area of rectangle given its width and height
    Parameters:
     width (float)
     height (float)
    Returns:
     area (float)
    '''
    area = width * height
    return area

cisc106.assertEqual(computeRectangleArea(0, 0), 0)
cisc106.assertEqual (computeRectangleArea(10, 10), 100)
cisc106.assertEqual (computeRectangleArea(1, 10), 10)
````
